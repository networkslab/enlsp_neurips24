```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
FILENAME = 'rouge_scores.csv'
```

```python
df = pd.read_csv(FILENAME)
#Column names are "rouge1", "rouge2", "rougeL", "label_length", "prediction_length","prompt_length"
rouge_type = "rouge1"
```

```python

#plot rouge_type vs label_length
plt.scatter(df["label_length"], df[rouge_type])
plt.xlabel("Label Length")
plt.ylabel(rouge_type)
plt.title(rouge_type + " vs Label Length")
#measure correlation
correlation = df[rouge_type].corr(df["label_length"])
#show correlation
plt.text(0, 0, "Correlation: " + str(correlation))
plt.show()

```

```python
#plot rouge_type vs prediction_length
plt.scatter(df["prediction_length"], df[rouge_type])
plt.xlabel("Prediction Length")
plt.ylabel(rouge_type)
plt.title(rouge_type + " vs Prediction Length")
correlation = df[rouge_type].corr(df["prediction_length"])
plt.text(0, 0, "Correlation: " + str(correlation))
plt.show()
```

```python
#plot rouge_type vs prompt_length
plt.scatter(df["prompt_length"], df[rouge_type])
plt.xlabel("Prompt Length")
plt.ylabel(rouge_type)
plt.title(rouge_type + " vs Prompt Length")
correlation = df[rouge_type].corr(df["prompt_length"])
plt.text(0, 0, "Correlation: " + str(correlation))
plt.show()
```
